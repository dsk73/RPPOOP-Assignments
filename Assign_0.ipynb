```python
class Polygon:
    def getArea(self):
        pass
    def getPerimeter(self):
        pass
```

```python
class Triangle(Polygon):
    def __init__(self, b, h):
        self.__base = b
        self.__height = h
        
    def setBase(self, b):
        self.__base = b
        return
    
    def setHeight(self, b):
        self.__height = b
        return
    
    def getBase(self):
        return self.__base
    
    def getHeight(self):
        return self.__height
        
    def getArea(self):
        area = 0.5 * self.__base * self.__height
        return area
```

```python
class Square(Polygon):
    def __init__(self, l):
        self.__length = l
        
    def setLength(self, l):
        self.__length = l
        return
    
    def getLength(self):
        return self.__length
        
    def getArea(self):
        area = self.__length * self.__length
        return area
    
    def getPerimeter(self):
        peri = 4*(self.__length)
        return peri
```

```python
class Rectangle(Polygon):
    def __init__(self, l, b):
        self.__length = l
        self.__breadth = b
        
    def setLength(self, l):
        self.__length = l
        return
    
    def setBreadth(self, b):
        self.__breadth = b
        return
    
    def getLength(self):
        return self.__length
    
    def getBreadth(self):
        return self.__breadth
        
    def getArea(self):
        area = self.__length * self.__breadth
        return area
    
    def getPerimeter(self):
        peri = 2*(self.__length + self.__breadth)
        return peri 
```

```python
def Menu():
    print("Choose the shape:")
    print("1. Traingle")
    print("2. Square")
    print("3. Rectangle")
    print("4. Exit")
```

```python
def TriangleMenu():
    print("Choose the action you want to perform")
    print("1. Get Base")
    print("2. Get Height")
    print("3. Change Base")
    print("4. Change Height")
    print("5. Get Area")
    print("6. Exit")
```

```python
def SquareMenu():
    print("Choose the action you want to perform")
    print("1. Get Side Length")
    print("2. Change Side Length")
    print("3. Get Area")
    print("4. Get Perimeter")
    print("5. Exit")
```

```python
def RectangleMenu():
    print("Choose the action you want to perform")
    print("1. Get Length")
    print("2. Get Breadth")
    print("3. Change Length")
    print("4. Change Breadth")
    print("5. Get Area")
    print("6. Get Perimeter")
    print("7. Exit")
```

```python
def displayTriangle():
    try:
        b = int(input("Enter the base of the triangle: "))
        h = int(input("Enter the height of the triangle: "))
        t1 = Triangle(b, h)

        while True:
            TriangleMenu()

            try:
                choice = int(input("Enter your choice: "))

                if choice == 1:
                    print(f"Base of the triangle is: {t1.getBase()}")
                elif choice == 2:
                    print(f"Height of the triangle is: {t1.getHeight()}")
                elif choice == 3:
                    b_new = int(input("Enter the new value of the base: "))
                    t1.setBase(b_new)
                elif choice == 4:
                    h_new = int(input("Enter the new value of the height: "))
                    t1.setHeight(h_new)
                elif choice == 5:
                    print(f"Area of the triangle is: {t1.getArea()}")
                elif choice == 6:
                    print("Exiting...")
                    return
                else:
                    print("Invalid choice!")
            except ValueError:
                print("Invalid input! Please enter a valid integer choice.")

            print()
    except ValueError:
        print("Invalid input! Please enter valid integer values for the base and height.")
```

```python
def displaySquare():
    try:
        l = int(input("Enter the length of the side of the square: "))
        c1 = Square(l)

        while True:
            SquareMenu()

            try:
                choice = int(input("Enter your choice: "))

                if choice == 1:
                    print(f"Length of the side of the square is: {c1.getLength()}")
                elif choice == 2:
                    l_new = int(input("Enter the new value of the side: "))
                    c1.setLength(l_new)
                elif choice == 3:
                    print(f"Area of the square is: {c1.getArea()}")
                elif choice == 4:
                    print(f"Perimeter of the square is: {c1.getPerimeter()}")
                elif choice == 5:
                    print("Exiting...")
                    return
                else:
                    print("Invalid choice!")
            except ValueError:
                print("Invalid input! Please enter a valid integer choice.")

            print()
    except ValueError:
        print("Invalid input! Please enter a valid integer value for the side length.")
```

```python
def displayRectangle():
    try:
        l = int(input("Enter the length of the rectangle: "))
        b = int(input("Enter the breadth of the rectangle: "))
        r1 = Rectangle(l, b)

        while True:
            RectangleMenu()

            try:
                choice = int(input("Enter your choice: "))

                if choice == 1:
                    print(f"Length of the rectangle is: {r1.getLength()}")
                elif choice == 2:
                    print(f"Breadth of the rectangle is: {r1.getBreadth()}")
                elif choice == 3:
                    l_new = int(input("Enter the new value of length: "))
                    r1.setLength(l_new)
                elif choice == 4:
                    b_new = int(input("Enter the new value of breadth: "))
                    r1.setBreadth(b_new)
                elif choice == 5:
                    print(f"Area of the rectangle is: {r1.getArea()}")
                elif choice == 6:
                    print(f"Perimeter of the rectangle is: {r1.getPerimeter()}")
                elif choice == 7:
                    print("Exiting...")
                    return
                else:
                    print("Invalid choice!")
            except ValueError:
                print("Invalid input! Please enter a valid integer choice.")

            print()
    except ValueError:
        print("Invalid input! Please enter valid integer values for length and breadth.")
```
