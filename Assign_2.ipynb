```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def get_length(self):
        return (self.x ** 2 + self.y ** 2) ** 0.5

    def __add__(self, *others):
        result = Vector(self.x, self.y)
        for other in others:
            if isinstance(other, Vector):
                result.x += other.x
                result.y += other.y
            else:
                raise TypeError("Unsupported operand type for +")
        return result

    def __sub__(self, *others):
        result = Vector(self.x, self.y)
        for other in others:
            if isinstance(other, Vector):
                result.x -= other.x
                result.y -= other.y
            else:
                raise TypeError("Unsupported operand type for -")
        return result

    def __str__(self):
        return f"({self.x}, {self.y})"

```

```python
class SpecialVector(Vector):
    def __init__(self, x, y, name):
        super().__init__(x, y)
        self.name = name

    def __add__(self, *others):
        result = Vector(self.x, self.y)
        for other in others:
            if isinstance(other, Vector):
                result.x -= other.x
                result.y -= other.y
            else:
                raise TypeError("Unsupported operand type for +")
        return result

    def __sub__(self, *others):
        result = Vector(self.x, self.y)
        for other in others:
            if isinstance(other, Vector):
                result.x += other.x
                result.y += other.y
            else:
                raise TypeError("Unsupported operand type for -")
        return result

    def __str__(self):
        return f"({-self.x}, {-self.y}) => {self.name}"
```

```python
def Menu():
    print("Choose the type of polymorphism:")
    print("1. Compile Time")
    print("2. Run Time")
    print("3. Exit")
```

```python
def CompileTime():
    print("Compile Time Polymorphism")
    print("1. Operator Overloading")
    print("2. Method Overloading")
    print("3. Exit")

    choice = int(input("Enter your choice: "))
    try:
        if choice == 1:
            v1 = Vector(1, 2)
            v2 = Vector(3, 4)
            print("v1 : ", v1)
            print("v2 : ", v2)
            print("v1 + v2 : ",v1 + v2)
            print("v1 - v2 : ",v1 - v2 )
            
        elif choice == 2:             
            v1 = Vector(1, 2)
            v2 = Vector(3, 4)
            v3 = Vector(5, 6)
            print("v1 : ", v1)
            print("v2 : ", v2)
            print("v3 : ", v3)
            print("v1 + v2 : ",v1 + v2)
            print("v1 + v2 + v3 : ",v1 + v2 + v3)
            print("v1 - v2 : ",v1 - v2 )
            print("v1 - v2 - v3 : ",v1 - v2 - v3)
        
        elif choice == 3:
            print("Exiting...")

        else:
            print("Invalid input! Please enter a valid choice.")

    except:
        print("Invalid input! Please enter a valid integer choice.")
```

```python
def RunTime():
    print("Run Time Polymorphism")
    print("1. Method Overriding")
    print("2. Exit")

    choice = int(input("Enter your choice: "))
    try:
        if choice == 1 :
            v1 = SpecialVector(1, 2, "Inversed v1")
            v2 = SpecialVector(3, 4, "Inversed v2")
            print("v1 : ", v1)
            print("v2 : ", v2)
            print("v1 + v2 : ",v1 + v2)
            print("v1 - v2 : ",v1 - v2 )

        elif choice == 2:
            print("Exiting...")
            
        else:
            print("Invalid input! Please enter a valid choice.")

    except:
        print("Invalid input! Please enter a valid integer choice.")
```

```python
while True:
    Menu()

    try:
        choice = int(input("Enter your choice: "))

        if choice == 1:
            CompileTime()
        elif choice == 2:
            RunTime()        
        elif choice == 3:
            print("Exiting...")
            print("Thank you.")
            break
        else:
            print("Invalid choice!")
    except ValueError:
        print("Invalid input! Please enter a valid integer choice.")

    print()
```
